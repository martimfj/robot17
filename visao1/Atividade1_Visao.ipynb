1. O histograma não aproveita bem as faixas de cores. As cores presentes na imagem estão entre 0 e 70 na escala RGB.

```python
x0 = np.amin(rintin)
y0 = 0 #minimo
x1 = np.amax(rintin)
y1 = 255 #max
m = (y1 - y0)/(x1 - x0) #coef angular
img_eq = rintin
```

```python
for i in range (len(img_eq)):
    for j in range (len(img_eq[i])):
        for w in range (len(img_eq[i][j])):
            img_eq[i][j][w] = m*(img_eq[i][j][w] - x0)
plt.imshow(img_eq)
```
